- **1)** Faça o gráfico da Renda [dos jogos do campeonato brasileiro de 2018](../dados/tabelaBrasileirao2018.csv) em função do público de cada jogo. Calcule a correlação entre essas grandezas.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/Daniellucasgonzaga/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')
df.plot.scatter(x='Público', y='Renda (R$)', figsize = (20,10)) # plota grafico de dispersão
cor = df['Público'].corr(df['Renda (R$)'])                      # calcula a correlação
print(f'A correlação entre o público e a renda é de {cor}')     # printa  na tela o valor da correlação
```
