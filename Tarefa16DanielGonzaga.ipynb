**1)** Faça uma função que receba um vetor **t**,um valor **A**, um valor **b** e um valor **c** e retorne um outro vetor **x**.

$$x(t)  = A\sqrt{1+bt^2} + c$$

Além disso a função deve plotar o gráfico de **x** em função de **t**.

Teste para 
- *a)* t indo de 0 a 10 com 1000 pontos, A = 2, b = 0.5, c = 0; 
- *b)* t indo de 0 a 15 com intervalo de 0.2 entre os valores do vetor, A = 10, b = 0.2, c = 1
- *c*) t indo de -0,5 a 0,5 com 500 pontos, A = -3, b = -1.5, c = -10; 

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao1(A, b, t, c):                # define a função e suas variáveis 
  f = A * np.sqrt(1+(b * (t**2))) + c   # monta equação
  return f                              # retorna a varaável f para ser usada fora da função

# criando as variáveis do exercício a
t1 = np.linspace(0, 10, 1000 ) 
A1 = 2
b1 = 0.5
c1 = 0

# criando as variáveis do exercício b                                                               
t2 = np.arange(0, 15.2, 0.2)
A2 = 10
b2 = 0.2
c2 = 1

# criando as variáveis do exercício c
t3 = np.linspace(-0.5, 0.5, 500)
A3 = -3
b3 = -1.5
c3 = -10

# Plotando os 3 gráficos em uma figura só
plt.figure(figsize=(20, 10))
plt.plot(funcao1(A1, b1, t1, c1), t1, label = 'exercício a')
plt.plot(funcao1(A2, b2, t2, c2), t2, label = 'exercício b')
plt.plot(funcao1(A3, b3, t3, c3), t3, label = 'exercício c')
plt.grid(True)
plt.ylabel('X(t)')
plt.xlabel('x')
plt.legend()

```

**3)** Faça uma função que retorne a renda média nos jogos em que determinado time seja o time da casa. As entradas devem ser o **DataFrame do Pandas** com os jogos e o **nome do time mandante**. 

Para testar a função, use o arquivo [com todos os resultados do campeonato Brasileiro de futebol de 2018](../dados/tabelaBrasileirao2018.csv) (dados obtidos [desta](<https://pt.wikipedia.org/wiki/Resultados_do_Campeonato_Brasileiro_de_Futebol_de_2018_-_S%C3%A9rie_A_(primeiro_turno)>) e [desta](<https://pt.wikipedia.org/wiki/Resultados_do_Campeonato_Brasileiro_de_Futebol_de_2018_-_S%C3%A9rie_A_(segundo_turno)>) página da Wikipedia) e escolha 4 times diferentes para saber quala renda média por jogo.

Dica: para usar o valor de uma variável (que tenha o nome var, por exemplo) dentro da função `query`, usar @var.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Daniellucasgonzaga/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')

# define a função e separa as tabelas em relação aos mandantes escolhidos
def renda_media():
  cor = df.query("Mandante == 'Corinthians'")
  crz = df.query("Mandante == 'Cruzeiro'")
  flu = df.query("Mandante == 'Fluminense'")
  inter = df.query("Mandante == 'Internacional'")
# calcula a média da renda para cada mandante
  med1 = cor['Renda (R$)'].median()
  med2 = crz['Renda (R$)'].median()
  med3 = flu['Renda (R$)'].median()
  med4 = inter['Renda (R$)'].median()
# mostra a média arrecadada para cada mandante
  print(f'A renda média quando o Corinthians foi o mandante foi de R$ {med1}')
  print(f'A renda média quando o Cruzeiro foi o mandante foi de R$ {med2}')
  print(f'A renda média quando o Fluminense foi o mandante foi de R$ {med3}')
  print(f'A renda média quando o Internacional foi o mandante foi de R$ {med4}')

renda_media() # retorna a função definida


```

**4)** Dados valores **x1**, **y1**, **x2**, **y2**, sendo esses valores das coordenadas $(x_1, y_1)$ e $(x_2, y_2)$ de dois pontos. 

Fazer uma função que encontre a inclinação da reta **m** e o ponto **b** em que a reta cruza o eixo y da reta $y = mx+b$ que passa pelos dois pontos. Além disso deve ser feito o gráfico da reta encontrada com **N** pontos, além de mostrar os dois pontos dados como entrada, indicados com marcadores quadrados.

Essa função deve retornar **m** e **b** e receber como parâmetros **x1**, **y1**, **x2**, **y2** e **N**.

Dica: a inclinação de uma reta, dados dois pontos é:

$$m = \frac{y_2-y_1}{x_2-x_1} $$

e o ponto que cruza o eixo y é:

$$b = y_1 - mx_1$$

Teste para: 

- a) x1 = -19, y1 = 2, x2 = 10, y2 = -10, N = 1000
- b) x1 = -2, y1 = 8, x2 = 20, y2 = 43, N = 2000
- c) x1 = 7, y1 = 13, x2 = 1, y2 = 2, N = 500

```python
import numpy as np
import matplotlib.pyplot as plt

def reta(x1, y1, x2, y2, x0): # define as variáveis da função
  m = (y2 - y1) / (x2 - x1)   # calcula o valor de m
  b = y1 - (m * x1)           # calcula o valor de b
  f = (m * x0) + b            # calcula a função
  return f                    # retorna a função para usarmos fora do def

# exercício a
xa0 = np.arange(0, 1001, 1)
xa1 = -19
ya1 = 2
xa2 = 10
ya2 = -10

# exercício b
xb0 = np.arange(0, 2001, 1)
xb1 = -2
yb1 = 8
xb2 = 20
yb2 = 43

# exercício c
xc0 = np.arange(0, 501, 1)
xc1 = 7
yc1 = 13
xc2 = 1
yc2 = 2

plt.figure()
plt.plot(xa0, reta(xa1, ya1, xa2, ya2, xa0))                                   # plota a reta
plt.scatter(xa1, ya1, label = '( -19,2 )', color = 'r', marker = 's', s = 100) # plota ponto inicial
plt.legend()                                                                   # adiciona legenda
plt.title('exercício a')                                                       # adiciona título
plt.xlabel('x')                                                                # nomeia o eixo x
plt.ylabel('f(x)')                                                             # nomeia o eixo y
plt.show()


plt.figure()
plt.plot(xb0, reta(xb1, yb1, xb2, yb2, xb0))
plt.scatter(xb1, yb1, label = '( -2,8 )', color = 'r', marker = 's', s = 100)
plt.legend()
plt.title('exercício b')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.figure()
plt.plot( xc0, reta(xc1, yc1, xc2, yc2, xc0))
plt.scatter(xc1, yc1, label = '( 7,13 )', color = 'r', marker = 's', s = 100)
plt.legend()
plt.title('exercício c')
plt.xlabel('x')
plt.ylabel('f(x)')
```
