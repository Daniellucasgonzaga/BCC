```python
#1) tensão elétrica medida em um circuito:
import matplotlib.pyplot as plt  
import numpy as np

t = np.arange(0,10.05,0.05)                                                      #Delimita o intervalo de operação do gráfico e o espaçamento dos valores
e = np.exp(-0.5*t)                                                               #Eleva o numero de Euler a cada número descrito do arange
pi = np.pi                                                                       #atribui a "pi" o seu valor numerico

v = e*(np.cos(2*pi*0.8*t))                                                       #Atribui os valores a função 

plt.figure(figsize=(10,8))                                                       #Plota a figura aonde será implementado o gráfico
plt.plot(t,v)                                                                    #Plota o gráfico
plt.xlabel('tempo (s)')                                                          #Nomeia o eixo X
plt.ylabel('tensão (V)')                                                         #Nomeia o eixo Y
plt.title('Grafico tensão elétrica')                                             #Adiciona um título ao gráfico
plt.show()                                                                       #Permite mostrar o grafico plotado ao executar o programa



```

```python
#2) Faça um gráfico da equivalência da temperatura em Fahrenheits em função da temperatura em Celsius:
import matplotlib.pyplot as plt  
import numpy as np

c = np.arange(-20,100,1.2)                                                       #Cria um conjunto universo para as variaveis solicitadas no exercicio
f = ((9/5)*c)+32                                                                 #Aplica a formula dada no exercicio

plt.figure(figsize=(10,8))                                                       #Plota a figura aonde será implementado o gráfico e ajusta seu tamanho
plt.plot(c,f)                                                                    #Plota o gráfico
plt.xlabel('°C')                                                                 #Nomeia o eixo X
plt.ylabel('F')                                                                  #Nomeia o eixo Y
plt.title('Grafico Celcius para Farenheit')                                      #Adiciona um título ao gráfico
plt.grid()                                                                       #Adiciona grade ao grafico
plt.show()                                                                       #Permite mostrar o grafico plotado ao executar o programa

```

```python
#3) crescimento populacional:
import matplotlib.pyplot as plt  
import numpy as np

a = np.arange(1994,2023,1)                                                       #Delimita o conjunto universo sendo de 1994 a 2022 com 1 ano de espaçamento
e = np.exp((a-1994)/40)                                                          #Eleva o numero de Euler a equação descrita pela formula
p = 800000*e                                                                     #Aplica a formua dada pelo exercício

plt.figure(figsize=(10,8))                                                       #Plota a figura aonde será implementado o gráfico e ajusta seu tamanho
plt.plot(a,p)                                                                    #Plota o gráfico
plt.xlabel('ano')                                                                #Nomeia o eixo X
plt.ylabel('Habitantes')                                                         #Nomeia o eixo Y
plt.title('Grafico habitantes por ano')                                          #Adiciona um título ao gráfico
plt.grid()                                                                       #Adiciona grade ao grafico
plt.show()                                                                       #Permite mostrar o grafico plotado ao executar o programa


```
