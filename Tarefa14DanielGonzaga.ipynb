**1)** Para este item você deverá usar o arquivo [com dados de força e torque medidos no chão enquanto uma pessoa tenta ficar em pé durante um minuto](dados/balance.csv).

- Faça o gráfico do torque na direção y (My [Nm]) em função do tempo (Time [s]).

- Mostre na tela a média e o desvio-padrão do torque na direção y.

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('https://raw.githubusercontent.com/Daniellucasgonzaga/BasesComputacionais/master/dados/balance.csv')
# plotando o gráfico 
df1.plot (kind='line', x='Time[s]', y='My[Nm]', figsize = (20,10))   
# calculando a média
md = np.mean(df1['My[Nm]'])
# calculando o desvio padrão
dp = df1['My[Nm]'].std()
# mostrando os valores da média e desvio padrão, com 4 casa decimais
print(f'A média do torque na direção y é de: {round(md,4)} Nm')
print(f'O desvio padrão do torque na direção y é de: {round(dp,4)}')
```

**2)**  Para este item você deverá usar o arquivo com a [taxa de inflação mensal (IGP-DI) no Brasil de fevereiro de 1944 a junho de 2019](dados/inflacaoMensal.csv) (dados disponibilizados por Ipeadata). 

- Calcule a mediana da inflação mensal nos meses de março.

- Calcule a média da inflação mensal nos meses de março.

- Calcule o desvio-padrão da inflação mensal nos meses de março.

- Faça o histograma da inflação mensal nos meses de março.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df2 = pd.read_csv('https://raw.githubusercontent.com/Daniellucasgonzaga/BasesComputacionais/master/dados/inflacaoMensal.csv')
df3 = df2.query("Mês == 3") # faz uma nova tabela somente com o mês de março

med1 = df3['Inflação'].median() # calcula a mediana da coluna inflação
print(f'A mediana dos valores de inflação mensal é de: {med1}') # mostra na tela a mediana 
md1 = np.mean(df3['Inflação']) # calcula a inflação média
print(f'A média dos valores de inflação mensal é de: {round(md1,4)}') # mostra a média na tela
dp1 = df3['Inflação'].std() # Calcula o desvio padrão
print(f'O desvio padrão dos valores de inflação mensal é de: {round(dp1,4)}') # mostra o desvio padrão na tela

df3.hist(column = 'Inflação', bins = 50, edgecolor='black', figsize = (25,10)) # plota o histograma
plt.title('Inflação mensal em março') # edita título
plt.xlabel('Inflação em março') # título no eixo x
plt.ylabel('Frequência absoluta') # título no eixo y
plt.show()
```

**3)**  Para este item você deverá usar o arquivo com a [taxa de inflação mensal (IGP-DI) no Brasil de fevereiro de 1944 a junho de 2019](dados/inflacaoMensal.csv) (dados disponibilizados por Ipeadata). 

- Calcule a mediana da inflação mensal nos meses de março a partir de 1995.

- Calcule a média da inflação mensal nos meses de março a partir de 1995.

- Calcule o desvio-padrão da inflação mensal nos meses de março a partir de 1995.

- Faça o histograma da inflação mensal nos meses de março a partir de 1995.

- Coloque um comentário no seu script comentando a razão da diferença entre a média e a mediana ser alta no segundo item e baixa no terceiro item.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df4 = pd.read_csv('https://raw.githubusercontent.com/Daniellucasgonzaga/BasesComputacionais/master/dados/inflacaoMensal.csv')
df5 = df4.query("Ano >= 1995 & Mês == 3") # faz uma tabela com as restrições dadas no exercício

med2 = df5['Inflação'].median() # calcula a mediana
print(f'A mediana dos valores de inflação mensal a partir de março de 1995 é de: {med2}') # mostra o valor da mediana na tela
md2 = np.mean(df5['Inflação']) # calcula a média
print(f'A média dos valores de inflação mensal a partir de março de 1995 é de: {round(md2,4)}') # mostra o valor da média na tela
dp2 = df5['Inflação'].std() # calcula o desvio padrão
print(f'O desvio padrão dos valores de inflação mensal a partir de março de 1995 é de: {round(dp2,4)}') # mostra o desvio padrão na tela

df5.hist(column = 'Inflação', bins = 30, edgecolor='black', figsize = (25,10)) # plota o histograma
plt.title('Inflação mensal em março') # edita título
plt.xlabel('Inflação desde março de 1995') # título no eixo x
plt.ylabel('Frequência absoluta') # título no eixo y
plt.show()
```

A razão de a diferença entre a média e a mediana ser alta no segundo item e baixa no terceiro item se dá porque houve uma restrição nos dados do dataframe, que ao invés de serem considerados todo o universo (inflação em março em todos os anos colotados) restringimos para aprenas a partir de 1995.

**4)** Use a sequência de instruções abaixo para gerar N números aleatórios entre 0 e 1. 

`rng = np.random.default_rng()`  
`x = rng.uniform(size=N)`

Faça o histograma e calcule (e mostre na tela) a média dos valores gerados.

Faça isso para a) N = 100, N = 1000, c) N = 10000, d) N = 100000

```python
import numpy as np
import matplotlib.pyplot as plt

# para N = 100
N1 = 100 
rng1 = np.random.default_rng()
x1 = rng1.uniform(size = N1)
plt.hist(x1, 20, rwidth = 0.9)                 # plota o histograma
plt.title('N = 100', fontsize = 20)            # adiciona um título
plt.ylabel('Frequência Absoluta')              # nomeia o eixo y
plt.show()
md1 = np.mean(x1)                              # calcula a média
print(f'A média para N=100 é: {round(md1,4)}') # mostra o valor da média na tela

# para N = 1000
N2 = 1000
rng2 = np.random.default_rng()
x2 = rng2.uniform(size = N2)
plt.hist(x2, 20, rwidth = 0.9)
plt.title('N = 1000', fontsize = 20)
plt.ylabel('Frequência Absoluta')
plt.show()
md2 = np.mean(x2)
print(f'A média para N=1000 é: {round(md2,4)}')

# para N = 10000
N3 = 10000
rng3 = np.random.default_rng()
x3 = rng3.uniform(size = N3)
plt.hist(x3, 20, rwidth = 0.9)
plt.title('N = 10000', fontsize = 20)
plt.ylabel('Frequência Absoluta')
plt.show()
md3 = np.mean(x3)
print(f'A média para N=10000 é: {round(md3,4)}')

# para N = 100000
N4 = 100000
rng4 = np.random.default_rng()
x4 = rng4.uniform(size = N4)
plt.hist(x1, 20, rwidth = 0.9)
plt.title('N = 100000', fontsize = 20)
plt.ylabel('Frequência Absoluta')
plt.show()
md4 = np.mean(x4)
print(f'A média para N=100000 é: {round(md4,4)}')
```
