**1)** O índice de massa corporal (IMC) de uma pessoa é calculado por:
    
$$IMC = \frac{m}{h^2} $$
    
em que $m$ é a massa corporal, em kg, e $h$ é a altura, em metros.

O IMC tem a seguinte classificação:

- menor que 17 kg/m$^2$ -- Muito abaixo do peso	
- 17 a 18,5 kg/m$^2$ -- Abaixo do peso	
- 18,5 a 25 kg/m$^2$ -- Peso normal
- 25 a 30 kg/m$^2$ -- Acima do peso	
- 30 a 35 kg/m$^2$ -- Obesidade Grau I	
- 35 a 40 kg/m$^2$ -- Obesidade Grau II	
- maior que 40 kg/m$^2$ -- Obesidade Grau III	

Escreva um código em Python para calcular o IMC e mostrar a classificação do IMC. Teste para a) $m = 52$ kg e $h = 1,58$ m, b) $m = 83$ kg e $h = 1,75$ m, c) seu peso e altura.

```python
m = 52
h = 1.58

imc = m/(h**2)

print('Para uma pessoa com 52Kg e 1,58m')   # Exibe o texto ao rodar o programa
if imc < 17:
    print('Situação: Muito abaixo do peso') # Mostra o texto caso o imc < 17
elif imc < 18.5:
    print('Situação: Abaixo do peso')       # Mostra o texto caso o imc < 18,5
elif imc < 25:
    print('Situação: Peso normal')          # Mostra o texto caso o imc < 25
elif imc < 30:
    print('Situação: Acima do peso')        # Mostra o texto caso o imc < 30
elif imc < 35:
    print('Situação: Obesidade grau I')     # Mostra o texto caso o imc < 35
elif imc < 40:
    print('Situação: Obesidade grau II')    # Mostra o texto caso o imc < 40
elif imc >= 40:
    print('Situação: Obesidade grau III')   # Mostra o texto caso o imc >= 40
```

**2)** A acidez $A(x)$ de uma solução de hidróxido de magnésio em ácido clorídrico, sob certas condições experimentais, é dada pela equação:

$$A(x) = x^3 + 3x^2 − 54$$
na qual $x$ é a concentração de íons hidrônio. 

Pede-se:

1) Gere o gráfico de $A(x)$ em função de $x$ para 0 $\leqslant x \leqslant 8$.

2) Determine a concentração $x$ do íon de hidrônio que resulta em solução saturada (i.e., com acidez nula). Acrescente uma instrução que gere um ponto vermelho no gráfico correspondente à acidez nula da solução.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange (0, 8.5, 0.5)                                            # Define o conjunto que será usado em x
y = ((x**3) + (3*(x**2)) - 54)                                         # Aplicação da fórmula


plt.plot(x, y,)                                                        # Plota o gráfico da função
plt.xlabel('concentração íons hidrônio')                               # Nomeia o eixo x
plt.ylabel('Acidez')                                                   # Nomeia o eixo y
plt.title('Grafico acidez de uma solução de Mg em HCl')                # Título da tabela
plt.plot(x[y==0],y[y==0],'ro')                                         # Plota um ponto no gráfico aonde x = 0
plt.annotate('Acidez nula',(x[y==0], y[y==0]), xytext=(0,500),         # adiciona uma anotação ao gráfico
             arrowprops=dict(facecolor='r', connectionstyle='Angle3')) # Adiciona a seta ao gáfico
plt.grid()                                                             # Plota uma grade no gráfico


```

**3)**  Faça o gráfico de:
    $$f(x)=|x−2|+|2x+1|−x−6$$
    
 Faça o gráfico com $x$ variando de -10 a 10.
 
 Mostre o menor valor de $x$ tal que $f(x)>0$ e $x>0$.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange (-10, 11, 1)                               # Intervalo de atuação e variação do gráfico
y = (np.abs(x-2)) + (np.abs((2*x)+1)) - x - 6            # Aplicação da fórmula

fig, ax = plt.subplots (figsize = (15,15))               # cria um subplot para o grafico com a classe axes
plt.plot(x, y, marker='o', color='red')                  # Plota o grafico de y em função de x           
plt.xlabel('X')                                          # Nomeia eixo x
plt.ylabel('Y')                                          # Nomeia eixo y
plt.title('Grafico função modular')                      # Título do gráfico
plt.annotate('f(x)>0 e x>0',(4, 1), xytext=(-2.5,20),    # adiciona um texto ao grafico
             arrowprops=dict(connectionstyle='Angle3'))  # configura a seta do grafico
for x1, y1 in zip (x, y):                                # concatena os valores de x com os de y
    str_xy = '  ({}, {})'.format(x1,y1)                  # formata a string que mostra os pares ordenados
    ax.annotate (str_xy, xy = (x1, y1), fontsize=10)     # anota os pares ordenados concatenados antertiormente
plt.grid()
```

**4)** Faça o gráfico da seguinte função:

$$f(x) = x^2 – \mathrm{sen}(0.784x^2) – 2$$

Mostre as raízes da função (i.e. os valores de $x$ para os quais $f(x) = 0$).

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-20, 20.1, 0.1)
y = (x**2) - np.sin(0.784 * (x**2)) -2


plt.figure(figsize=(10,10))
plt.plot(x, y,)            
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafico função modular')
plt.annotate('raiz da função',(0, 0), xytext=(0,200),    # adiciona um texto ao grafico
             arrowprops=dict(connectionstyle='Angle3'))  # configura a seta do grafico
plt.grid()
```
