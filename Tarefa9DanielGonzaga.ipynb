1) Para este item você deverá usar o arquivo com a porcentagem da população brasileira que é analfabeta, dividida por unidade da federação, de 1981 a 2014 (analfabetismo.csv, encontrado na pasta aula5/dados) (dados disponibilizados por Ipeadata).

Faça um gráfico da taxa de analfabetismo em São Paulo em função do tempo (anos).


```python
import pandas as pd
import matplotlib.pyplot as plt

analfabetismo = pd.read_csv ('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/analfabetismo.csv')

plt.figure(figsize=(20,10))
plt.plot(analfabetismo['Ano'], analfabetismo['São Paulo'],
         marker = 'o', 
         linestyle = '', 
         color = 'red', 
         markersize = 15)
plt.grid()
plt.xlabel('Ano')
plt.ylabel('Taxa de Analfabetismo')
plt.show()
```
